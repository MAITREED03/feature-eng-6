Q1. Pearson correlation coefficient is a measure of the linear relationship between two variables. Suppose
you have collected data on the amount of time students spend studying for an exam and their final exam
scores. Calculate the Pearson correlation coefficient between these two variables and interpret the result.

To calculate the Pearson correlation coefficient between the amount of time students spend studying for an exam and their final exam scores, the following steps are typically undertaken:

Organize the data: Gather the data for both variables (study time and exam scores) for each student.
Compute the means: Calculate the mean (average) of study time and exam scores.
Calculate the deviations: Find the deviations of each data point from their respective means for both variables.
Compute the product of deviations: Multiply the deviations for each data point and sum the products.
Calculate the standard deviations: Compute the standard deviations for both variables.
Determine the correlation coefficient: Divide the sum of the product of deviations by the product of the standard deviations of both variables.
Once the Pearson correlation coefficient is calculated, it will yield a value between -1 and 1:

A value close to 1 indicates a strong positive linear relationship, implying that as study time increases, exam scores tend to increase.
A value close to -1 indicates a strong negative linear relationship, suggesting that as study time increases, exam scores tend to decrease.
A value close to 0 indicates little to no linear relationship between the two variables.
After obtaining the Pearson correlation coefficient, interpretation should be based on the magnitude and sign of the coefficient, providing insights into the relationship between study time and exam scores for the given dataset.

Q4. A researcher is interested in examining the relationship between the number of hours individuals
spend watching television per day and their level of physical activity. The researcher collected data on
both variables from a sample of 50 participants. Calculate the Pearson correlation coefficient between
these two variables.

To calculate the Pearson correlation coefficient between the number of hours individuals spend watching television per day and their level of physical activity, the researcher needs to follow these steps:

Organize the data: Record the number of hours spent watching television per day and the corresponding level of physical activity for each participant in the sample.

Calculate the mean for each variable: Compute the mean (average) number of hours spent watching television per day and the mean level of physical activity in the sample.

Calculate the deviations from the mean: For each participant, subtract the mean number of hours spent watching television per day from their individual number of hours, and do the same for physical activity levels.

Multiply the corresponding deviations: Multiply the deviations obtained in step 3 for each participant.

Sum the products: Add up all the products obtained in step 4.

Calculate the standard deviation for each variable: Compute the standard deviation of the number of hours spent watching television per day and the standard deviation of the level of physical activity in the sample.

Multiply the standard deviations: Multiply the standard deviations obtained in step 6 for each variable.

Divide the sum of the products (step 5) by the product of the standard deviations (step 7).

Interpret the result: The Pearson correlation coefficient will range between -1 and 1. A coefficient close to 1 indicates a strong positive linear relationship between the variables (more hours of television watching associated with higher physical activity levels), a coefficient close to -1 indicates a strong negative linear relationship (more hours of television watching associated with lower physical activity levels), and a coefficient close to 0 indicates no linear relationship.

By following these steps, the researcher can accurately calculate the Pearson correlation coefficient between the number of hours individuals spend watching television per day and their level of physical activity in the sample.
